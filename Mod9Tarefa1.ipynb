### 1. Idade media das mães e dos pais por município (coluna munResNome)


```python
sinasc1 = sinasc_raw[['munResNome', 'IDADEMAE', 'IDADEPAI']]
sinasc1.groupby('munResNome').mean()
```

### 2. Peso médio dos bebes por sexo que nasceram no dia do seu aniversário por faixas de escolaridade mae
Ex: Você, aluna(o), nasceu no dia 10/01, então você precisa filtrar o conjunto de dados nessa data e calcular o peso médio dos bebês de cada sexo por faixa de escolaridade da mãe.

```python
sinasc2 = sinasc_raw[['PESO', 'SEXO', 'DTNASC', 'ESCMAE']]
```

```python
sinasc2.loc[sinasc2['DTNASC'] == '2019-06-15'].groupby(['SEXO', 'ESCMAE'])[['PESO']].mean()
```

### 3. Qual o municipio que nasceu menos bebe em 2019?
    - qual a idade media, maxima, minima das maes nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?

```python
#Qual municipio nasceu menos bebe em 2019 = Castanheiras
sinasc_raw['munResNome'].value_counts()
```

Como o municipio de Castanheiras é o com menos bebe nascidos em 2019 vamos utilizar ele.

```python
sinasc_mun_menos = sinasc_raw[sinasc_raw['munResNome'] == 'Castanheiras'].groupby('munResNome')[['IDADEMAE', 'IDADEPAI']]
```

```python
sinasc_mun_menos.agg([('idade minima', 'min'), ('idade média', 'mean'), ('idade máxima', 'max')])
```

### 4. Qual o municipio que nasceu mais bebe no mês de março?
    - qual a quantidade de filhos vivos media, maxima, minima nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?



Municipio que mais nasceram bebes foi o de porto velho.

```python
sinasc_raw['DTNASC'] = pd.to_datetime(sinasc_raw['DTNASC'])
marco = sinasc_raw[sinasc_raw['DTNASC'].dt.month == 3]
```

```python
marco.groupby('munResNome')['PARTO'].count().sort_values(ascending=False).head(5)
```

```python
marco.loc[marco['munResNome'] == 'Porto Velho'].groupby('munResNome')[['QTDFILVIVO', 'IDADEPAI',
                                                                      'IDADEMAE']].agg(['min',
                                                                                       'max', 'mean'])
```
